```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import logging
import names
from random import randint
```

```python
# locate chrome driver for browser
s=Service(r'C:\Users\jakub\OneDrive\Plocha\chromedriver.exe') 
browser = webdriver.Chrome(service=s)

# define url to open
url="https://forms.gle/uJRzBGC7Z91fHvtHA"
browser.get(url)
browser.maximize_window() # max window of browser
```

```python
# find/locate elements in form by Xpath
first_name = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[1]/div/div/div[2]/div/div[1]/div/div[1]/input'
last_name = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input'
email = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/input'
sex_male = '//*[@id="i17"]/div[3]/div'
sex_female = '//*[@id="i20"]/div[3]/div'
submit = '//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span/span'

# create test screnarios/data
scenarios = [
    {
        'first_name': 'Jon',
        'last_name': 'Snow',
        'email': 'jon.snow@gmail.com',
        'sex': 1
    },
    {
        'first_name': 'Will',
        'last_name': 'Smith',
        'email': 'will.smith@gmail.com',
        'sex': 1
    },
    {
        'first_name': 'Daenerys',
        'last_name': 'Targaryen',
        'email': 'daenerys.targaryen@gmail.com',
        'sex': 2
    }
]

# read data from scenarios and fill them into form
for scenario in scenarios:
    fn_input = browser.find_element(By.XPATH, first_name)
    fn_input.click
    fn_input.send_keys(scenario['first_name'])

    ln_input = browser.find_element(By.XPATH, last_name)
    ln_input.click
    ln_input.send_keys(scenario['last_name'])

    email_inp = browser.find_element(By.XPATH, email)
    email_inp.click
    email_inp.send_keys(scenario['email'])

    if scenario['sex'] == 1:
        gender = browser.find_element(By.XPATH, sex_male)
        gender.click()
    else:
        gender = browser.find_element(By.XPATH, sex_female)
        gender.click()

    # submit form
    submit_btn = browser.find_element(By.XPATH, submit)
    submit_btn.click()

    # send another response
    another_resp = browser.find_element(By.LINK_TEXT, 'Odeslat další odpověď')
    another_resp.click()


```

```python
# Automate whole process and randomize data for responses

# create function for fill the form
def fillform_random(iterations):
    # locate chrome driver for browser
    s=Service(r'C:\Users\jakub\OneDrive\Plocha\chromedriver.exe') 
    browser = webdriver.Chrome(service=s)

    # define url to open
    url="https://forms.gle/uJRzBGC7Z91fHvtHA"
    browser.get(url)
    browser.maximize_window() # max window of browser
    first_name = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[1]/div/div/div[2]/div/div[1]/div/div[1]/input'
    last_name = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input'
    email = '//*[@id="mG61Hd"]/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/input'
    sex_male = '//*[@id="i17"]/div[3]/div'
    sex_female = '//*[@id="i20"]/div[3]/div'
    submit = '//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span/span'

    for i in range(0,iterations):
        # generate random data for form based on gender
        gender_data = randint(1,2)
        if gender_data == 1:
            first_name_data = names.get_first_name(gender='male')
        else:
            first_name_data = names.get_first_name(gender='female')
        last_name_data = names.get_last_name()
        email_data = f"{first_name_data.lower()}.{last_name_data.lower()}@gmail.com"

        fn_input = browser.find_element(By.XPATH, first_name)
        fn_input.click
        fn_input.send_keys(first_name_data)

        ln_input = browser.find_element(By.XPATH, last_name)
        ln_input.click
        ln_input.send_keys(last_name_data)

        email_inp = browser.find_element(By.XPATH, email)
        email_inp.click
        email_inp.send_keys(email_data)

        if gender_data == 1:
            gender = browser.find_element(By.XPATH, sex_male)
            gender.click()
        else:
            gender = browser.find_element(By.XPATH, sex_female)
            gender.click()

        # submit form
        submit_btn = browser.find_element(By.XPATH, submit)
        submit_btn.click()

        # send another response
        another_resp = browser.find_element(By.LINK_TEXT, 'Odeslat další odpověď')
        another_resp.click()





```
